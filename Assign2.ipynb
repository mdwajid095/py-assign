Write a Python Program to implement your own my_reduce() function which works exactly like python's built-in function reduce()

```python
l=[1,2,3,4,5]
add = lambda a, b:a+b
mul = lambda a, b:a*b
div = lambda a, b:a/b

#for comparing the results
from functools import reduce
print(reduce(add, l))
print(reduce(mul, l)) 
print(reduce(div, l)) 
```

```python
def my_reduce(fun, l):
    n=l[0]
    for i in range(2,len(l)+1):
        n= fun(n, i)
    return n
    
print(my_reduce(add, l)) 
print(my_reduce(mul, l)) 
print(my_reduce(div, l)) 
```

Write a Python program to implement your own my_filter() function which works exactly like Python's built-in function filter()

```python
l=[1,2,3,4,5,8,9,20,25]
def even(i) :
    if i%2==0:
        return True

import math
def perfectSqr(i):
    if math.floor(i**0.5)==i**0.5:
        return True
    
#for comparing the results
print('Even Number in list(l) are:')
print(list(filter(even, l)))
print('Perfect square number in list are :') 
print(list(filter(perfectSqr, l))) 
```

```python
def my_filter(fun, l):
    a=[]
    for i in l:
        if fun(i):
            a.append(i)
    return a
    
print('Even Number in list(l) are:')
print(my_filter(even, l))
print('Perfect square number in list are :') 
print(my_filter(perfectSqr, l))
```

Implement a function longestWord() that takes a list of words and returns the longest one.

```python
def longestWord(word) :
    l_word= ' '
    for i in word:
        if len(i) > len(l_word) :
            l_word = i
    return l_word

word=['I','am','wajid','a','intuitive','and','executive','guy']
print(longestWord(word))
```

Write a Python Program(with class concepts) to find the area of the triangle using the below
formula.
area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
Function to take the length of the sides of triangle from user should be defined in the parent
class and function to calculate the area should be defined in subclass.

```python
class triangle:
    def __init__(self, hypo, per, base) :
        self.hypo= hypo
        self.per= per
        self.base= base

#sub-class of triangle class
class area(triangle):
    def cal_area(self) :
        s=(self.hypo+self.per+self.base)/2
        a = (s*(s-self.hypo)*(s-self.per)*(s-self.base)) ** 0.5
        return a

obj=area(4,5,6)
obj.cal_area() 
    
```

Write a function filter_long_words() that takes a list of words and an integer n and returns the list
of words that are longer than n. 

```python
def filter_long_word(word, n):
    l_word= [ ]
    for i in word:
        if len(i)>n:
            l_word.append(i) 
    return l_word

word=['I','am','wajid','a','intuitive','and','executive','guy']
filter_long_word(word, 3)
```

Write a Python program using function concept that maps list of words into a list of integers representing the lengths of the corresponding words.

Hint: ​If a list [ ab,cde,erty] is passed on to the python function output should come as [2,3,4]
Here 2,3 and 4 are the lengths of the words in the list.

```python
def word2len(word):
    lenWord = [ ]
    for i in word:
        lenWord.append(len(i)) 
    return lenWord
    
word =['i', 'am', 'wajid']
word2len(word) 
```

Write a Python function which takes a character (i.e. a string of length 1) and returns True if it is
a vowel, False otherwise.

```python
def isVowel(a):
    vowel=['A','E','I','O','U','a','e','i','o','u']
    if a in vowel:
        return True
    return False

print(isVowel('A'))
print(isVowel('k'))
print(isVowel('i')) 
```
